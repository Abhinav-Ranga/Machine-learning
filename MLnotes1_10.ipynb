# Challenges in machine learning

1 : Data Collection

Gathering relevant data for training a machine learning model.

2 : Insufficient Data / Labelled Data

Having too little data or too few labeled examples, which makes it hard for the model to learn properly.

3 : Non-Representative Data

Data that does not reflect real-world situations, causing poor performance on new data.

4 : Poor Quality Data

Data containing errors, missing values, duplicates, or noise.

5 : Irrelevant Features

Input variables that do not help the model make accurate predictions.

6 : Overfitting

When a model memorizes the training data and performs poorly on unseen data.

7 : Underfitting

When a model is too simple and fails to learn the underlying patterns in the data


8 : Software Integration

Difficulty in incorporating the ML model into existing software or systems.

9 : Offline Learning / Deployment

Challenges in updating or deploying models that are trained once and do not learn continuously from new data.

10 : Cost Involved

The expense of collecting data, training models, maintaining infrastructure, and deploying ML systems.

# Machine learning development life cycle


MLDLC : it's a set of guidelines that we have to follow while making machine learning based software product




image.png

1. Frame the Problem

      framing the problem before starting

    like

    what is the cost

    who are our clients

    what is the exact problem

    simple word : What are we trying to predict or fix?



2. Gathering Data

    Collecting relevant information and raw datasets from various sources like databases, APIs, web scraping, or surveys.

    simple words: Finding and downloading the raw data.

3. Data Preprocessing

    Cleaning up the raw data by handling missing values, removing duplicate data, fixing errors, and converting text or categories into numbers.

    simple words: Fixing the messy data so the computer can understand it.

4. Exploratory Data Analysis (EDA)

    we try to study relationship b/w input and output

    we do experiments with our data

    we extract relationships in data



5. Feature Engineering and Selection

     Creating new, useful data columns (features) from existing ones and picking only the most important columns that will actually help the model learn.

    simple words: Choosing and designing the best inputs for your AI.

6. Model Training, Evaluation, and Tuning

    Feeding the prepared data into a machine learning algorithm so it can learn (Training), checking how accurate it is using test data (Evaluation), and adjusting its settings to get the best performance (Tuning).

    In simple words: Teaching the AI, test-scoring it, and making it smarter.

7. Model Deployment

    Taking the finalized, trained model and launching it into a live, real-world application (like an app or website) where real users can interact with it.



8. Testing

   Continuously checking the deployed model with real-world user data to make sure it works perfectly under heavy pressure, doesn't crash, and gives accurate live results.

    ex = u send new features to some trusted customer and get get their feedback and analyze if u want to improve something or not


9. Optimize

    Monitoring the model's performance over time, updating it with fresh data, and fixing any slowdowns or drops in accuracy.

    before launching ur model on server .

    We take some steps

    ex:

    model backup

    data backup

    handline

    loadbalancing

    simple words: Keeping the AI updated so it stays fast and smart.
