# Encoding categorical variables



Encoding categorical is 2nd type of feature transformation 

categorical has two types 
(1) : Nominal = No relationship or order between categories  
(2) : ordinal =  there is relationship or order between categories 




* If input data has columns that can be encoded to numbers then we use ordinal encoding.

* Ordinal encoding class works on ordinal categorical data

```python

* Label Encoding
if target variable[output] has categorical data then we use label encoder.
it is an action or a process you perform on that data. It is a data preprocessing technique.
```

```python
import pandas as pd

# Direct Raw URL to the customer.csv file
url = "https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day26-ordinal-encoding/customer.csv"

# Importing the dataset
df = pd.read_csv(url)
```

```python
df.sample(5)
```

```python
df.iloc[0:,2:]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,2:4], df.iloc[:,-1],test_size = 0.2)
```

```python
x_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = [['Poor','Average', 'Good'],['School','UG','PG']])
```

```python
oe.fit(x_train)
# we fit only train  but transform both 
# coz fit just read the data doesn't do anything so if do this with test also then this training won't make sense
```

```python
x_train = oe.transform(x_train)
x_test  = oe.transform(x_test)
```

```python
x_train
```

```python
oe.categories
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
