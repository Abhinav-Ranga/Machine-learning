# Tensor
tensor is nothing but a container for storing numbers.

.sometimes we store characters,strings but in rare cases

image.png

```python
import numpy as np

 # (1): 0D tensor = where dimension is zero
#  ex = any scalar 5,3,4,etc

a = np.array(9)
a.ndim
```

```python
# (2) : 1D tensor

# where dimension is 1
# ex  a vector  = [1,2,3,4]

a = np.array([23,34,23])
a.ndim

# here the dimension of this tensor is 1 but dimension of a vector depends upon number of elements inside vector

# dimension of tensor is related to number of axis

# dimension = axis(Rank)
```

```python
# (3) : 2D tensor
# where dimensions are  2
# ex  a matrix

a = np.array([[1,2],[3,4]])
a.ndim



# in  simple words = 2D tensor is made up of multiple 1d tensors
# all rows in a matrix are  vectors
```

```python
# (4) : 3D tensor
# where dimensions are  3

a = np.array([[[1,2],[3,4],[5,6]]])
a.ndim



# in  simple words = 3D tensor is made up of multiple 2d tensors

# same for 4d it will be made up of multiple 3d tensors ...... ND
```
